```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pickle
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [13.5, 9.0]
plt.rcParams['font.size'] = 24.
```

```python
data_dict = {}
for file_name in os.listdir():
    if (('run-' in file_name) and 
        ('pickle' in file_name)):
        with open(file_name, 'rb') as pickle_file:
            data_dict[file_name] = pickle.load(pickle_file)
```

```python
data_dict['run-1.pickle'].keys()
```

# Learning Curves

```python
for i, ls in zip(range(1,14), ['--', '-', '-.', '-', ':', '-.', '--',
                              '-', '-.', '--', ':', '-.', ':']):
    data_pickle = 'run-' + str(i) + '.pickle'
    plt.plot(data_dict[data_pickle]['step_numbers'], 
         data_dict[data_pickle]['loss_function_values'], 
             lw = 3, ls = ls, 
             label = str(data_dict[data_pickle]['num_compressed_features']))

plt.xscale('log')
plt.yscale('log')
    
plt.title('Learning Curves vs Number of Outputs')
plt.xlabel('Step Number')
plt.ylabel('Loss Function')

plt.legend()
```

# Final Loss vs. Number of Compressed Features

```python
final_loss = []
num_compressed = []

for i in range(1,14):
    data_pickle = 'run-' + str(i) + '.pickle'
    final_loss.append(
        data_dict[data_pickle]['loss_function_values'][-1])
    num_compressed.append(
        data_dict[data_pickle]['num_compressed_features'])
    
plt.plot(num_compressed, final_loss)
plt.scatter(num_compressed, final_loss)

plt.xscale('log')
plt.yscale('log')
    
plt.title('Learning Curves vs Number of Outputs')
plt.xlabel('Number of Compressed Features')
plt.ylabel('Loss Function')

plt.legend()
```
